#Modificación de datos

En el primer analisis veo que la mayoria de datos aparecen como "object" e "int64". Necesito pasarlos a string y datetime.


##Funciones generales

```python
#Función para cambiar las fechas a Datetime
def transformar_fechas(df, columnas_fechas):
    df[columnas_fechas] = df[columnas_fechas].apply(pd.to_datetime)
    return df
```

```python
#Función para cambiar los Object que son cadenas a Strings
def transformar_cadenas(df, columnas_cadenas):
  for columna in columnas_cadenas:
    df[columna] = df[columna].astype("string")
  return df
```

```python
#Función para completar con la media
def completar_media(df, columnas_media):
    for columna in columnas_media:
        df[columna] = df[columna].fillna(df[columna].mean())
    return df
```

```python
#Función para completar con 0 más de un dato nulo
def fillna_cero(df, columnas):
    df[columnas] = df[columnas].fillna(0)
    return df
```

```python
#Aplico la función para cambiar los datos a strings
columnas_cadenas_ordenes=["order_id","customer_id","order_status"]
df_ecommerce_orders_dataset = transformar_cadenas(df_ecommerce_orders_dataset,columnas_cadenas_ordenes)
```

```python
#calculo el promedio de diferencia entre columnas de orden de compra y compra arpobada para poder completar los 160 datos nulos
diferencia_tiempo = (df_ecommerce_orders_dataset["order_approved_at"] - df_ecommerce_orders_dataset["order_purchase_timestamp"]).mean()
```

```python
#completo los nulos con el promedio de las fechas
df_ecommerce_orders_dataset["order_approved_at"]=df_ecommerce_orders_dataset["order_approved_at"].fillna(df_ecommerce_orders_dataset["order_purchase_timestamp"] + diferencia_tiempo)
```

Para poder rellenar los otros datos nulos que me quedan voy a definir una funcion que me ayude a sacar el promedio de dias que hay entre el momento de la compra y el momento busqueda y entrega. Para luego sumar esa cantidad de dias promedio a las fechas que tengo en la columna de "order_purchase_timestamp" y asi completar los datos nulos de las columnas "order_delivered_carrier_date" y "order_delivered_customer_date".

```python
def calcular_promedio_envio(df, columna_compra, columna_envio):
    tiempo_promedio_envio = (df[columna_envio] - df[columna_compra]).dt.days.mean()

    return tiempo_promedio_envio
```

```python
#Uso la función para calcular el tiempo promedio de busqueda del envio
promedio_envio = calcular_promedio_envio(df_ecommerce_orders_dataset, "order_purchase_timestamp", "order_delivered_carrier_date")
print("El tiempo promedio de envío es:", promedio_envio, "días")
```

```python
df_ecommerce_orders_dataset["order_delivered_carrier_date"] = df_ecommerce_orders_dataset["order_delivered_carrier_date"].fillna(df_ecommerce_orders_dataset["order_purchase_timestamp"] + pd.Timedelta(days=3))
```

```python
#Uso la función para calcular el tiempo promedio de entrega del envio
promedio_envio = calcular_promedio_envio(df_ecommerce_orders_dataset, "order_purchase_timestamp", "order_delivered_customer_date")
print("El tiempo promedio de entrega es:", promedio_envio, "días")
```

```python
df_ecommerce_orders_dataset["order_delivered_customer_date"] = df_ecommerce_orders_dataset["order_delivered_customer_date"].fillna(df_ecommerce_orders_dataset["order_purchase_timestamp"] + pd.Timedelta(days=12))
```

#Analisis de DataFrame

```python
#Voy a definir una función que me devuelva datos relevantes en base al dtype que encuentre
def analisis_data(df):
    print("*******")
    print("Análisis variables cuantitativas")
    columnas_numericas = df.select_dtypes(include=['int', 'float']).columns
    if not columnas_numericas.empty:
        print(df[columnas_numericas].describe())
    else:
        print("No hay columnas numéricas en el DataFrame.")

    print("*******")
    print("Análisis variables cualitativas")
    columnas_strings = df.select_dtypes(include=['string']).columns
    if not columnas_strings.empty:
        print(df[columnas_strings].describe())
    else:
        print("No hay columnas cualitativas en el DataFrame.")

    print("*******")
    print("Análisis variables de fecha y hora")
    columnas_datetime = df.select_dtypes(include=['datetime']).columns
    if not columnas_datetime.empty:
        print(df[columnas_datetime].describe())
    else:
        print("No hay columnas de fecha y hora en el DataFrame.")
```
